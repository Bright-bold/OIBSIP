## Feature Engineering

```python
# One-Hot Encoding

pd.get_dummies(data.Car_Name).astype(int)
```

```python
data= data.join(pd.get_dummies(data.Car_Name).astype(int)).drop(['Car_Name'],axis = 1)
data
```

```python
data= data.join(pd.get_dummies(data.Fuel_Type).astype(int)).drop(['Fuel_Type'],axis = 1)
data
```

```python
data= data.join(pd.get_dummies(data.Transmission).astype(int)).drop(['Transmission'],axis = 1)
data
```

```python
data= data.join(pd.get_dummies(data.Selling_type).astype(int)).drop(['Selling_type'],axis = 1)
data
```

### Prediction of selling price

```python
from sklearn.model_selection import train_test_split
X = data.drop(['Selling_Price'],axis =1)
y = data['Selling_Price']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2)
```

```python
train_df = X_train.join(y_train)
train_df
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()



reg.fit(X_train,y_train)
reg.score(X_train,y_train)
```

## Standardization of Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()



X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)


scaler.fit(X_train_s,y_train)
reg.fit(X_train_s,y_train)
```

```python
reg.score(X_train_s,y_train)
```

## Modelling using Random Regressor

```python
from sklearn.ensemble import RandomForestRegressor  # Use RandomForestClassifier for classification
from sklearn.model_selection import train_test_split


# Initialize the Random Forest Regressor
rf_reg = RandomForestRegressor(random_state=42)  # Use RandomForestClassifier for classification tasks

# Train the model
rf_reg.fit(X_train_s, y_train)

# Get the R-squared score on the training set
train_score = rf_reg.score(X_train_s, y_train)

# Get the R-squared score on the testing set
test_score = rf_reg.score(X_test_s, y_test)
train_score

```

```python
y_pred = rf_reg.predict(X_test_s)

comparison_df = pd.DataFrame({ "Predicted Selling_price": y_pred[:21],"Actual Selling_price": y_test[:21].values})

# Display the DataFrame in a clean format
print(comparison_df.to_string(index=False))
```
